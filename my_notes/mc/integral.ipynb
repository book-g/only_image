Here’s a **comprehensive list of integral calculus formulas**, systematically organized to cover basic to advanced cases, including definite/indefinite integrals, techniques, and special functions:

---

### **1. Basic Indefinite Integrals**
#### **Polynomials & Powers**
1. $$ \int x^n \, dx = \frac{x^{n+1}}{n+1} + C \quad (n \neq -1) $$
2. $$ \int \frac{1}{x} \, dx = \ln|x| + C $$
3. $$ \int e^x \, dx = e^x + C $$
4. $$ \int a^x \, dx = \frac{a^x}{\ln a} + C \quad (a > 0, a \neq 1) $$

#### **Trigonometric Functions**
5. $$ \int \sin x \, dx = -\cos x + C $$
6. $$ \int \cos x \, dx = \sin x + C $$
7. $$ \int \tan x \, dx = -\ln|\cos x| + C $$
8. $$ \int \cot x \, dx = \ln|\sin x| + C $$
9. $$ \int \sec x \, dx = \ln|\sec x + \tan x| + C $$
10. $$ \int \csc x \, dx = -\ln|\csc x + \cot x| + C $$
11. $$ \int \sec^2 x \, dx = \tan x + C $$
12. $$ \int \csc^2 x \, dx = -\cot x + C $$
13. $$ \int \sec x \tan x \, dx = \sec x + C $$
14. $$ \int \csc x \cot x \, dx = -\csc x + C $$

#### **Inverse Trigonometric**
15. $$ \int \frac{1}{\sqrt{1-x^2}} \, dx = \sin^{-1}x + C $$
16. $$ \int \frac{-1}{\sqrt{1-x^2}} \, dx = \cos^{-1}x + C $$
17. $$ \int \frac{1}{1+x^2} \, dx = \tan^{-1}x + C $$
18. $$ \int \frac{-1}{1+x^2} \, dx = \cot^{-1}x + C $$

#### **Hyperbolic Functions**
19. $$ \int \sinh x \, dx = \cosh x + C $$
20. $$ \int \cosh x \, dx = \sinh x + C $$
21. $$ \int \text{sech}^2 x \, dx = \tanh x + C $$
22. $$ \int \text{csch}^2 x \, dx = -\coth x + C $$

### **2. Definite Integrals**
1. **Fundamental Theorem of Calculus:**  
   $$ \int_a^b f(x) \, dx = F(b) - F(a), \text{ where } F'(x) = f(x). $$

2. **Properties:**  
   - $\int_a^a f(x) \, dx = 0$  
   - $\int_a^b f(x) \, dx = -\int_b^a f(x) \, dx$  
   - $\int_a^b [f(x) \pm g(x)] \, dx = \int_a^b f(x) \, dx \pm \int_a^b g(x) \, dx$  
   - $\int_a^b c f(x) \, dx = c \int_a^b f(x) \, dx$  

3. **Mean Value Theorem for Integrals:**  
   $$ \exists c \in [a, b] \text{ s.t. } f(c) = \frac{1}{b-a} \int_a^b f(x) \, dx. $$

---

### **3. Integration Techniques**
#### **Substitution (u-Sub)**
$$ \int f(g(x)) g'(x) \, dx = \int f(u) \, du, \quad u = g(x). $$  
**Example:**  
$\int 2x \cos(x^2) \, dx = \sin(x^2) + C$.

#### **Integration by Parts**
$$ \int u \, dv = uv - \int v \, du. $$  
**Example:**  
$\int x e^x \, dx = x e^x - e^x + C$.

#### **Partial Fractions**
For $\int \frac{P(x)}{Q(x)} \, dx$ (where $Q(x)$ is factorizable):  
Decompose into simpler fractions, e.g.,  
$$ \int \frac{1}{x^2-1} \, dx = \frac{1}{2} \ln\left|\frac{x-1}{x+1}\right| + C. $$

#### **Trigonometric Substitution**
1. $\sqrt{a^2 - x^2}$: Use $x = a \sin \theta$.  
2. $\sqrt{a^2 + x^2}$: Use $x = a \tan \theta$.  
3. $\sqrt{x^2 - a^2}$: Use $x = a \sec \theta$.  

**Example:**  
$\int \frac{1}{\sqrt{1-x^2}} \, dx = \sin^{-1}x + C$.

---

### **4. Improper Integrals**
1. **Infinite Limits:**  
   $$ \int_a^\infty f(x) \, dx = \lim_{b \to \infty} \int_a^b f(x) \, dx. $$  
2. **Discontinuities:**  
   $$ \int_a^b f(x) \, dx \text{ where } f \text{ is undefined at } c \in (a,b). $$

---

### **5. Special Integrals**
#### **Exponential & Logarithmic**
1. $$ \int \ln x \, dx = x \ln x - x + C $$  
2. $$ \int \frac{f'(x)}{f(x)} \, dx = \ln|f(x)| + C $$

#### **Gaussian Integral**
$$ \int_{-\infty}^\infty e^{-x^2} \, dx = \sqrt{\pi}. $$

#### **Beta & Gamma Functions**
1. Gamma: $\Gamma(n) = \int_0^\infty x^{n-1} e^{-x} \, dx \quad (n > 0)$.  
2. Beta: $B(p, q) = \int_0^1 x^{p-1} (1-x)^{q-1} \, dx = \frac{\Gamma(p)\Gamma(q)}{\Gamma(p+q)}$.

---

### **6. Reduction Formulas**
Recursive formulas for repeated integration:  
1. **Sine/Cosine Powers:**  
   $$ \int \sin^n x \, dx = -\frac{\sin^{n-1} x \cos x}{n} + \frac{n-1}{n} \int \sin^{n-2} x \, dx. $$  
2. **Secant Powers:**  
   $$ \int \sec^n x \, dx = \frac{\sec^{n-2} x \tan x}{n-1} + \frac{n-2}{n-1} \int \sec^{n-2} x \, dx. $$

---

### **7. Multivariable Integrals**
1. **Double Integrals:**  
   $$ \iint_R f(x,y) \, dA = \int_{a}^{b} \int_{c}^{d} f(x,y) \, dy \, dx. $$  
2. **Polar Coordinates:**  
   $$ \iint_R f(r,\theta) \, r \, dr \, d\theta. $$  
3. **Triple Integrals:**  
   $$ \iiint_V f(x,y,z) \, dV. $$

### **9. Applications**
1. **Area Under Curve:** $A = \int_a^b f(x) \, dx$.  
2. **Arc Length:** $L = \int_a^b \sqrt{1 + (f'(x))^2} \, dx$.  
3. **Surface Area:** $S = 2\pi \int_a^b f(x) \sqrt{1 + (f'(x))^2} \, dx$.  
4. **Volume of Revolution:**  
   - Disk: $V = \pi \int_a^b [f(x)]^2 \, dx$.  
   - Shell: $V = 2\pi \int_a^b x f(x) \, dx$.
