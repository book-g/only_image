Here’s a **comprehensive list of partial derivative formulas**, covering functions of one or multiple variables, chain rules, and advanced cases. I’ve organized them systematically to ensure nothing is missed.

---

### **1. Basic Partial Derivatives**
For a function $f(x, y, z, \dots)$:

1. **First-Order Partial Derivatives:**  
   - $\frac{\partial f}{\partial x}$: Derivative w.r.t. $x$ (treat other variables as constants).  
   - $\frac{\partial f}{\partial y}$: Derivative w.r.t. $y$ (similarly for $z$, etc.).  

   **Example:**  
   For $f(x, y) = x^2 y + \sin(y)$:  
   $$
   \frac{\partial f}{\partial x} = 2xy, \quad \frac{\partial f}{\partial y} = x^2 + \cos(y).
   $$

---

### **2. Higher-Order Partial Derivatives**
1. **Second-Order Partial Derivatives:**  
   - $\frac{\partial^2 f}{\partial x^2}$: Differentiate $\frac{\partial f}{\partial x}$ w.r.t. $x$ again.  
   - $\frac{\partial^2 f}{\partial y \partial x}$: Differentiate $\frac{\partial f}{\partial x}$ w.r.t. $y$ (mixed partial).  

   **Schwarz’s Theorem:** If $f$ is $C^2$ (continuous second derivatives), then:  
   $$
   \frac{\partial^2 f}{\partial x \partial y} = \frac{\partial^2 f}{\partial y \partial x}.
   $$

   **Example:**  
   For $f(x, y) = e^{xy}$:  
   $$
   \frac{\partial^2 f}{\partial x \partial y} = \frac{\partial}{\partial x} \left( x e^{xy} \right) = e^{xy} (1 + xy).
   $$

---

### **3. Chain Rules for Partial Derivatives**
#### **(a) Single-Variable Chain Rule (Composite Functions)**
If $z = f(x, y)$ and $x = x(t)$, $y = y(t)$:  
$$
\frac{dz}{dt} = \frac{\partial f}{\partial x} \frac{dx}{dt} + \frac{\partial f}{\partial y} \frac{dy}{dt}.
$$

#### **(b) Multivariable Chain Rule**
If $z = f(x, y)$ and $x = x(u, v)$, $y = y(u, v)$:  
$$
\frac{\partial z}{\partial u} = \frac{\partial f}{\partial x} \frac{\partial x}{\partial u} + \frac{\partial f}{\partial y} \frac{\partial y}{\partial u},  
$$  
$$
\frac{\partial z}{\partial v} = \frac{\partial f}{\partial x} \frac{\partial x}{\partial v} + \frac{\partial f}{\partial y} \frac{\partial y}{\partial v}.
$$

**Example:**  
For $z = x^2 y$, $x = uv$, $y = u + v$:  
$$
\frac{\partial z}{\partial u} = 2xy \cdot v + x^2 \cdot 1 = 2uv(u+v)v + u^2v^2.
$$

---

### **4. Implicit Differentiation (Multivariable)**
For $F(x, y, z) = 0$ defining $z$ implicitly as $z(x, y)$:  
$$
\frac{\partial z}{\partial x} = -\frac{\frac{\partial F}{\partial x}}{\frac{\partial F}{\partial z}}, \quad \frac{\partial z}{\partial y} = -\frac{\frac{\partial F}{\partial y}}{\frac{\partial F}{\partial z}}.
$$

**Example:**  
For $x^2 + y^2 + z^2 = 1$:  
$$
\frac{\partial z}{\partial x} = -\frac{2x}{2z} = -\frac{x}{z}.
$$

---

### **5. Gradient, Divergence, and Curl**
1. **Gradient (∇f):**  
   $$
   \nabla f = \left( \frac{\partial f}{\partial x}, \frac{\partial f}{\partial y}, \frac{\partial f}{\partial z} \right).
   $$

2. **Divergence (∇·F):**  
   For a vector field $\mathbf{F} = (P, Q, R)$:  
   $$
   \nabla \cdot \mathbf{F} = \frac{\partial P}{\partial x} + \frac{\partial Q}{\partial y} + \frac{\partial R}{\partial z}.
   $$

3. **Curl (∇×F):**  
   $$
   \nabla \times \mathbf{F} = \begin{vmatrix}
   \mathbf{i} & \mathbf{j} & \mathbf{k} \\
   \frac{\partial}{\partial x} & \frac{\partial}{\partial y} & \frac{\partial}{\partial z} \\
   P & Q & R
   \end{vmatrix}.
   $$

---

### **6. Taylor Series Expansion (Multivariable)**
For $f(x, y)$ near $(a, b)$:  
$$
f(x, y) \approx f(a, b) + \frac{\partial f}{\partial x}\big|_{(a,b)} (x-a) + \frac{\partial f}{\partial y}\big|_{(a,b)} (y-b) + \frac{1}{2!} \left( \frac{\partial^2 f}{\partial x^2} (x-a)^2 + 2 \frac{\partial^2 f}{\partial x \partial y} (x-a)(y-b) + \frac{\partial^2 f}{\partial y^2} (y-b)^2 \right).
$$

---

### **7. Special Cases**
#### **(a) Homogeneous Functions (Euler’s Theorem)**
If $f$ is homogeneous of degree $n$ (i.e., $f(tx, ty) = t^n f(x, y)$):  
$$
x \frac{\partial f}{\partial x} + y \frac{\partial f}{\partial y} = n f(x, y).
$$

#### **(b) Exact Differentials**
For $M dx + N dy$ to be exact:  
$$
\frac{\partial M}{\partial y} = \frac{\partial N}{\partial x}.
$$

---

### **8. Partial Derivatives of Common Functions**
| Function                     | Partial Derivatives                                                                 |
|------------------------------|------------------------------------------------------------------------------------|
| $f(x, y) = x^a y^b$      | $\frac{\partial f}{\partial x} = a x^{a-1} y^b$, $\frac{\partial f}{\partial y} = b x^a y^{b-1}$ |
| $f(x, y) = e^{ax + by}$  | $\frac{\partial f}{\partial x} = a e^{ax + by}$, $\frac{\partial f}{\partial y} = b e^{ax + by}$ |
| $f(x, y) = \ln(x^2 + y^2)$ | $\frac{\partial f}{\partial x} = \frac{2x}{x^2 + y^2}$, $\frac{\partial f}{\partial y} = \frac{2y}{x^2 + y^2}$ |

---

### **9. Key Observations**
1. **Clairaut’s Theorem:** Mixed partials are equal if continuous ($\frac{\partial^2 f}{\partial x \partial y} = \frac{\partial^2 f}{\partial y \partial x}$).  
2. **Notation:** $\frac{\partial}{\partial x}$ (partial) vs. $\frac{d}{dx}$ (total derivative).
