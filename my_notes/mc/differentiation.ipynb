Here’s a comprehensive list of **differentiation formulas**, covering basic rules, trigonometric functions, exponentials, logarithms, and inverse functions. These are essential for computing derivatives in calculus.

---

### **1. Basic Differentiation Rules**
1. **Constant Rule:**  
   $$
   \frac{d}{dx} [c] = 0 \quad (c \text{ is a constant})
   $$

2. **Power Rule:**  
   $$
   \frac{d}{dx} [x^n] = n x^{n-1}
   $$

3. **Constant Multiple Rule:**  
   $$
   \frac{d}{dx} [c f(x)] = c f'(x)
   $$

4. **Sum/Difference Rule:**  
   $$
   \frac{d}{dx} [f(x) \pm g(x)] = f'(x) \pm g'(x)
   $$

5. **Product Rule:**  
   $$
   \frac{d}{dx} [f(x) g(x)] = f'(x) g(x) + f(x) g'(x)
   $$

6. **Quotient Rule:**  
   $$
   \frac{d}{dx} \left[ \frac{f(x)}{g(x)} \right] = \frac{f'(x) g(x) - f(x) g'(x)}{[g(x)]^2}
   $$

7. **Chain Rule:**  
   $$
   \frac{d}{dx} [f(g(x))] = f'(g(x)) \cdot g'(x)
   $$

---

### **2. Trigonometric Functions**
1. $$
   \frac{d}{dx} [\sin x] = \cos x
   $$
2. $$
   \frac{d}{dx} [\cos x] = -\sin x
   $$
3. $$
   \frac{d}{dx} [\tan x] = \sec^2 x
   $$
4. $$
   \frac{d}{dx} [\cot x] = -\csc^2 x
   $$
5. $$
   \frac{d}{dx} [\sec x] = \sec x \tan x
   $$
6. $$
   \frac{d}{dx} [\csc x] = -\csc x \cot x
   $$

---

### **3. Exponential & Logarithmic Functions**
1. $$
   \frac{d}{dx} [e^x] = e^x
   $$
2. $$
   \frac{d}{dx} [a^x] = a^x \ln a \quad (a > 0)
   $$
3. $$
   \frac{d}{dx} [\ln x] = \frac{1}{x} \quad (x > 0)
   $$
4. $$
   \frac{d}{dx} [\log_a x] = \frac{1}{x \ln a} \quad (a > 0, a \neq 1)
   $$

---

### **4. Inverse Trigonometric Functions**
1. $$
   \frac{d}{dx} [\sin^{-1} x] = \frac{1}{\sqrt{1 - x^2}} \quad (|x| < 1)
   $$
2. $$
   \frac{d}{dx} [\cos^{-1} x] = -\frac{1}{\sqrt{1 - x^2}} \quad (|x| < 1)
   $$
3. $$
   \frac{d}{dx} [\tan^{-1} x] = \frac{1}{1 + x^2}
   $$
4. $$
   \frac{d}{dx} [\cot^{-1} x] = -\frac{1}{1 + x^2}
   $$
5. $$
   \frac{d}{dx} [\sec^{-1} x] = \frac{1}{|x| \sqrt{x^2 - 1}} \quad (|x| > 1)
   $$
6. $$
   \frac{d}{dx} [\csc^{-1} x] = -\frac{1}{|x| \sqrt{x^2 - 1}} \quad (|x| > 1)
   $$

---

### **5. Hyperbolic Functions**
1. $$
   \frac{d}{dx} [\sinh x] = \cosh x
   $$
2. $$
   \frac{d}{dx} [\cosh x] = \sinh x
   $$
3. $$
   \frac{d}{dx} [\tanh x] = \text{sech}^2 x
   $$
4. $$
   \frac{d}{dx} [\coth x] = -\text{csch}^2 x
   $$
5. $$
   \frac{d}{dx} [\text{sech } x] = -\text{sech } x \tanh x
   $$
6. $$
   \frac{d}{dx} [\text{csch } x] = -\text{csch } x \coth x
   $$

---

### **6. Implicit Differentiation**
- Used when $y$ is defined implicitly as a function of $x$.  
  **Example:**  
  For $x^2 + y^2 = 1$, differentiate both sides w.r.t. $x$:  
  $$
  2x + 2y \frac{dy}{dx} = 0 \implies \frac{dy}{dx} = -\frac{x}{y}
  $$

---

### **7. Higher-Order Derivatives**
- The **second derivative**:  
  $$
  \frac{d^2y}{dx^2} = \frac{d}{dx} \left( \frac{dy}{dx} \right)
  $$
- The **n-th derivative**:  
  $$
  \frac{d^n y}{dx^n} = \frac{d}{dx} \left( \frac{d^{n-1} y}{dx^{n-1}} \right)
  $$

### **When to Use Which Rule?**
- **Power Rule:** For polynomials ($x^n$).  
- **Product/Quotient Rule:** When functions are multiplied/divided.  
- **Chain Rule:** For composite functions ($f(g(x))$).  
- **Implicit Differentiation:** When $y$ is not isolated.

### **2. Commonly Missed Formulas**
#### **(a) Inverse Hyperbolic Functions**
1. $$
   \frac{d}{dx} [\sinh^{-1} x] = \frac{1}{\sqrt{x^2 + 1}}
   $$
2. $$
   \frac{d}{dx} [\cosh^{-1} x] = \frac{1}{\sqrt{x^2 - 1}} \quad (x > 1)
   $$
3. $$
   \frac{d}{dx} [\tanh^{-1} x] = \frac{1}{1 - x^2} \quad (|x| < 1)
   $$
4. $$
   \frac{d}{dx} [\coth^{-1} x] = \frac{1}{1 - x^2} \quad (|x| > 1)
   $$

#### **(b) Parametric Differentiation**
- If $x = f(t)$ and $y = g(t)$, then:  
  $$
  \frac{dy}{dx} = \frac{dy/dt}{dx/dt}
  $$

#### **(c) Logarithmic Differentiation**
- For functions like $f(x)^{g(x)}$ or complex products:  
  $$
  \text{Let } y = f(x), \text{ then } \ln y = \ln f(x) \implies \frac{y'}{y} = \frac{f'(x)}{f(x)}.
  $$

#### **(d) Partial Derivatives**
- For multivariable functions (e.g., $f(x,y)$):  
  $$
  \frac{\partial f}{\partial x}, \frac{\partial f}{\partial y}.
  $$

#### **(e) Leibniz Rule for Integrals**
- Differentiating under the integral sign:  
  $$
  \frac{d}{dx} \left( \int_{a(x)}^{b(x)} f(x,t) \, dt \right) = f(x,b(x)) \cdot b'(x) - f(x,a(x)) \cdot a'(x) + \int_{a(x)}^{b(x)} \frac{\partial f}{\partial x} \, dt.
  $$

### **3. Special Cases & Niche Formulas**
#### **(a) Absolute Value**
$$
\frac{d}{dx} |x| = \frac{x}{|x|} = \text{sgn}(x) \quad (x \neq 0).
$$

#### **(b) Piecewise Functions**
- Differentiate each piece separately (e.g., $f(x) = \begin{cases} x^2 & x \geq 0 \\ -x & x < 0 \end{cases}$).

#### **(c) Higher-Order Derivatives of Implicit Functions**
- Use implicit differentiation recursively (e.g., for $\frac{d^2y}{dx^2}$).

#### **(d) Derivatives of Integrals (Fundamental Theorem of Calculus)**
$$
\frac{d}{dx} \left( \int_a^x f(t) \, dt \right) = f(x).
$$
