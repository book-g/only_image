# L3v3 : Instantaneous Acceleration in 2D

Here is a **step-by-step algorithm** to compute and visualize **2D motion**, showing how **position ‚Üí velocity ‚Üí acceleration** evolves:

---

## üîÅ Step-by-Step Algorithm: 2D Motion Kinematics (x(t), y(t))

### ‚úÖ **1. Define Coordinate System & Initial Functions**

* Define unit vectors:

  $$
  \hat{i}, \hat{j} \quad \text{(along x and y axes)}
  $$
* Define the position functions:

  $$
  x(t),\quad y(t)
  $$
* These represent the location of the particle at any time `t`.

---

### ‚úÖ **2. Compute Velocity Vector**

* Velocity is the **first derivative** of position:

  $$
  \vec{v}(t) = \frac{dx(t)}{dt} \hat{i} + \frac{dy(t)}{dt} \hat{j}
  $$
* Call components:

  $$
  v_x(t) = \frac{dx(t)}{dt}, \quad v_y(t) = \frac{dy(t)}{dt}
  $$

---

### ‚úÖ **3. Compute Acceleration Vector**

* Acceleration is the **derivative of velocity**:

  $$
  \vec{a}(t) = \frac{dv_x(t)}{dt} \hat{i} + \frac{dv_y(t)}{dt} \hat{j}
  $$
* Or:

  $$
  a_x(t) = \frac{d^2x(t)}{dt^2},\quad a_y(t) = \frac{d^2y(t)}{dt^2}
  $$

---

### ‚úÖ **4. Describe Velocity Change (Œîv)**

* Choose two times: $t_1$ and $t_2$
* Compute velocity vectors:

  $$
  \vec{v}_1 = \vec{v}(t_1),\quad \vec{v}_2 = \vec{v}(t_2)
  $$
* Translate vectors to **same origin** (parallel transport) to visualize:

  $$
  \Delta \vec{v} = \vec{v}_2 - \vec{v}_1
  $$

---

### ‚úÖ **5. Compute Instantaneous Acceleration**

* Let $\Delta t = t_2 - t_1$, then:

  $$
  \vec{a}(t) = \lim_{\Delta t \to 0} \frac{\Delta \vec{v}}{\Delta t}
  $$
* Use derivatives (from Step 3) as the instantaneous acceleration.

---

### ‚úÖ **6. Compute Magnitude of Vectors**

* Speed (magnitude of velocity):

  $$
  |\vec{v}| = \sqrt{v_x^2 + v_y^2}
  $$
* Acceleration magnitude:

  $$
  |\vec{a}| = \sqrt{a_x^2 + a_y^2}
  $$

---

### ‚úÖ **7. Visualize the Motion**

* Plot trajectory: $(x(t), y(t))$
* At multiple time steps:

  * Draw **velocity vector** at each point
  * Translate earlier velocity to new point and draw **Œîv**
  * Optionally draw acceleration vector

---

## üßÆ Summary Table

| Quantity         | Formula                             | Vector Form                                         |
| ---------------- | ----------------------------------- | --------------------------------------------------- |
| Position         | $x(t), y(t)$                        | $\vec{r}(t) = x(t)\hat{i} + y(t)\hat{j}$            |
| Velocity         | $v_x = dx/dt,\ v_y = dy/dt$         | $\vec{v}(t) = v_x(t)\hat{i} + v_y(t)\hat{j}$        |
| Acceleration     | $a_x = d^2x/dt^2,\ a_y = d^2y/dt^2$ | $\vec{a}(t) = a_x(t)\hat{i} + a_y(t)\hat{j}$        |
| Œîv               | $\vec{v}(t+\Delta t) - \vec{v}(t)$  | Translate vectors to same origin before subtracting |
| Speed            | $\sqrt{v_x^2 + v_y^2}$              | Scalar (magnitude)                                  |
| Acceleration mag | $\sqrt{a_x^2 + a_y^2}$              | Scalar (magnitude)                                  |

---

## üß™ Example

Let:

* $x(t) = 5t$
* $y(t) = 3t - 0.5 \cdot 9.8t^2$

Then:

* $v_x = 5,\ v_y = 3 - 9.8t$
* $a_x = 0,\ a_y = -9.8$
* $|\vec{v}| = \sqrt{5^2 + (3 - 9.8t)^2}$

---



```python
import numpy as np
import matplotlib.pyplot as plt

# Time settings
t = np.linspace(0, 2, 100)
dt = t[1] - t[0]

# Define position functions
def x(t): return 5 * t
def y(t): return 3 * t - 0.5 * 9.8 * t**2

# Velocity components (first derivative)
vx = np.gradient(x(t), dt)
vy = np.gradient(y(t), dt)

# Acceleration components (second derivative)
ax = np.gradient(vx, dt)
ay = np.gradient(vy, dt)

# Magnitudes
speed = np.sqrt(vx**2 + vy**2)
acceleration_magnitude = np.sqrt(ax**2 + ay**2)

# Choose a few sample indices to plot vectors
indices = [10, 30, 50, 70, 90]

# Plotting
plt.figure(figsize=(10, 8))
plt.plot(x(t), y(t), label='Trajectory', color='black')

# Plot velocity and acceleration vectors at selected points
for i in indices:
    plt.quiver(x(t)[i], y(t)[i], vx[i], vy[i], color='blue', scale=50, label='Velocity' if i == indices[0] else "")
    plt.quiver(x(t)[i], y(t)[i], ax[i], ay[i], color='red', scale=100, label='Acceleration' if i == indices[0] else "")

plt.title("2D Motion: Velocity and Acceleration Vectors")
plt.xlabel("x(t)")
plt.ylabel("y(t)")
plt.legend()
plt.grid(True)
plt.axis('equal')
plt.show()
```
