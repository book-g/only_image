# Introduction to Acceleration



### Algorithm: Understanding and Defining Acceleration

1. **Recognize velocity as rate of change of position**
2. **Understand acceleration as rate of change of velocity**
3. **Observe object motion (e.g., thrown upwards)**
4. **Note velocity changes continuously during motion**
5. **Question if acceleration is constant or varies over time**
6. **Examine object at highest point (velocity = 0)**
7. **Apply Newtonâ€™s second law: $F = m a$**
8. **Identify gravitational force acting continuously**
9. **Conclude acceleration is non-zero despite zero velocity**
10. **Summarize: acceleration exists throughout motion due to gravity**
11. **Mathematically define acceleration as derivative of velocity:**
    $a(t) = \frac{dv}{dt}$
12. **Express acceleration as second derivative of position:**
    $a(t) = \frac{d^2x}{dt^2}$



```python
import sympy as sp

# Step 1: Define symbolic variable for time t
t = sp.symbols('t')

# Step 2: Define position function x(t) (example: quadratic motion)
x = t**2  # Position function (can be changed as needed)

# Step 3: Calculate velocity as first derivative of position
v = sp.diff(x, t)

# Step 4: Calculate acceleration as derivative of velocity
a = sp.diff(v, t)

# Step 5: Print the results
print(f"Position x(t): {x}")
print(f"Velocity v(t) = dx/dt: {v}")
print(f"Acceleration a(t) = dv/dt = d^2x/dt^2: {a}")
```

```python
# Define position function x(t)
def position(t):
    return t**2  # example position function

# Numerical derivative function
def derivative(f, t, dt=1e-6):
    return (f(t + dt) - f(t)) / dt

# Numerical second derivative function
def second_derivative(f, t, dt=1e-6):
    return (f(t + dt) - 2*f(t) + f(t - dt)) / (dt**2)

# Choose a time value
t = 2.0

# Calculate velocity and acceleration numerically
velocity = derivative(position, t)
acceleration = second_derivative(position, t)

print(f"At t = {t}:")
print(f"Position: {position(t)}")
print(f"Velocity (numerical derivative): {velocity}")
print(f"Acceleration (numerical second derivative): {acceleration}")
```

### Problem:

Position function:

$$
x(t) = 5 \times 10^{-10} \times t^2 \quad \text{(meters)}, \quad t \text{ in attoseconds } (10^{-18} s)
$$

Find:

* Instantaneous velocity $v(t)$ at $t = 10$ attoseconds.
* Instantaneous acceleration $a(t)$ at $t = 10$ attoseconds.

---

### Step 1: Velocity $v(t)$

Velocity is the first derivative of position:

$$
v(t) = \frac{dx}{dt} = \frac{d}{dt}\left(5 \times 10^{-10} t^2\right)
$$

Using the power rule:

$$
v(t) = 2 \times 5 \times 10^{-10} \times t = 10^{-9} \times t
$$

At $t = 10$ attoseconds:

$$
v(10) = 10^{-9} \times 10 = 10^{-8} \text{ meters per attosecond}
$$

---

### Step 2: Acceleration $a(t)$

Acceleration is the derivative of velocity (second derivative of position):

$$
a(t) = \frac{dv}{dt} = \frac{d}{dt}(10^{-9} t) = 10^{-9}
$$

This means acceleration is constant:

$$
a(t) = 10^{-9} \text{ meters per attosecond squared}
$$

At $t = 10$ attoseconds (same value, constant):

$$
a(10) = 10^{-9} \text{ meters per attosecond}^2
$$

---

### Step 3: Unit conversion to SI units

* 1 attosecond = $10^{-18}$ seconds
* Velocity in $m/s$:

$$
v = 10^{-8} \frac{\text{meters}}{\text{attosecond}} = 10^{-8} \times \frac{1}{10^{-18}} = 10^{10} \, m/s
$$

* Acceleration in $m/s^2$:

$$
a = 10^{-9} \frac{m}{\text{attosecond}^2} = 10^{-9} \times \frac{1}{(10^{-18})^2} = 10^{-9} \times 10^{36} = 10^{27} \, m/s^2
$$

---

### Summary:

| Quantity         | Value (attosecond units)     | Converted value (SI units) |
| ---------------- | ---------------------------- | -------------------------- |
| Velocity $v$     | $10^{-8} \, m/\text{atto}$   | $10^{10} \, m/s$           |
| Acceleration $a$ | $10^{-9} \, m/\text{atto}^2$ | $10^{27} \, m/s^2$         |

---

### Note:

* Velocity exceeds the speed of light (\~$3 \times 10^8 m/s$), so this is a **theoretical/simplified calculation ignoring physical constraints**.
* The acceleration is extremely large due to the very small time unit squared in the denominator.
* This example illustrates differentiation and unit conversion principles.
