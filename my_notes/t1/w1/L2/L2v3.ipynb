## Question: Acceleration from Position (Worked Example)

A runner travels along the x-axis. At time $t = 0$, the runner is at the origin. The $x$-component of the runner's position with respect to the origin is given by the equation:

$$
x(t) = b t^2
$$

where $b$ is a positive constant.

---

### (a) What are the units of the constant $b$? Express your answer in terms of meters (m) and seconds (s).

---

### (b) Find the $x$-component of the runner's velocity as a function of time, $v(t)$.

---

### (c) Find the $x$-component of the runner's acceleration as a function of time, $a(t)$.

---

### (d) Plot the functions $x(t)$, $v(t)$, and $a(t)$ versus time $t$.

---


Got it! Let’s solve this **step-by-step** exactly as the question asks:

---

### Given:

$$
x(t) = b t^2
$$

where $b$ is a positive constant.

---

### (a) Units of $b$:

* Position $x(t)$ has units of meters (m).
* Time $t$ has units of seconds (s).
* Since $x(t) = b t^2$, the units of $b$ must satisfy:

$$
\text{units of } b \times (\text{seconds})^2 = \text{meters}
$$

$$
\Rightarrow [b] = \frac{\text{meters}}{\text{seconds}^2} = \text{m/s}^2
$$

**Answer:**
Units of $b$ are **meters per second squared (m/s²)**.

---

### (b) Velocity $v(t)$:

Velocity is the derivative of position with respect to time:

$$
v(t) = \frac{dx}{dt} = \frac{d}{dt} (b t^2) = 2 b t
$$

---

### (c) Acceleration $a(t)$:

Acceleration is the derivative of velocity with respect to time:

$$
a(t) = \frac{dv}{dt} = \frac{d}{dt} (2 b t) = 2 b
$$

---


```python
import numpy as np
import matplotlib.pyplot as plt

# Given constant b (positive)
b = 1.0  # m/s^2, you can change this value

# Time array from 0 to 10 seconds
t = np.linspace(0, 10, 500)

# Position x(t)
x = b * t**2

# Velocity v(t)
v = 2 * b * t

# Acceleration a(t)
a = 2 * b * np.ones_like(t)

# Plotting
plt.figure(figsize=(10, 8))

plt.subplot(3, 1, 1)
plt.plot(t, x, label='Position $x(t) = b t^2$')
plt.ylabel('Position (m)')
plt.legend()
plt.grid(True)

plt.subplot(3, 1, 2)
plt.plot(t, v, label='Velocity $v(t) = 2 b t$', color='orange')
plt.ylabel('Velocity (m/s)')
plt.legend()
plt.grid(True)

plt.subplot(3, 1, 3)
plt.plot(t, a, label='Acceleration $a(t) = 2 b$', color='green')
plt.xlabel('Time (s)')
plt.ylabel('Acceleration (m/s²)')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```
