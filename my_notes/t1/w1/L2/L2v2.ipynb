### Algorithm: From Position Vector to Acceleration (with intermediate concepts)

1. **Position vector $\mathbf{r}(t)$:** 


   * The location of the object at time $t$.
   * Example: 3D $\mathbf{r}(t) = x(t)\hat{i} + y(t)\hat{j} + z(t)\hat{k}$.
   * for 1D  $\mathbf{r}(t) = x(t)\hat{i}$

2. **Displacement vector $\Delta \mathbf{r}$:**

   * Change in position between times $t$ and $t + \Delta t$:

     $$
     \Delta \mathbf{r} = \mathbf{r}(t + \Delta t) - \mathbf{r}(t)
     $$

3. **Average velocity $\mathbf{v}_{avg}$:**

   * Displacement divided by the time interval $\Delta t$:

     $$
     \mathbf{v}_{avg} = \frac{\Delta \mathbf{r}}{\Delta t} = \frac{\mathbf{r}(t + \Delta t) - \mathbf{r}(t)}{\Delta t}
     $$

4. **Instantaneous velocity $\mathbf{v}(t)$:**

   * Limit of average velocity as $\Delta t \to 0$, i.e., the derivative of position:

     $$
     \mathbf{v}(t) = \lim_{\Delta t \to 0} \frac{\mathbf{r}(t + \Delta t) - \mathbf{r}(t)}{\Delta t} = \frac{d\mathbf{r}}{dt}
     $$

5. **Change in velocity $\Delta \mathbf{v}$:**

   * Change in velocity between times $t$ and $t + \Delta t$:

     $$
     \Delta \mathbf{v} = \mathbf{v}(t + \Delta t) - \mathbf{v}(t)
     $$

6. **Average acceleration $\mathbf{a}_{avg}$:**

   * Change in velocity divided by $\Delta t$:

     $$
     \mathbf{a}_{avg} = \frac{\Delta \mathbf{v}}{\Delta t} = \frac{\mathbf{v}(t + \Delta t) - \mathbf{v}(t)}{\Delta t}
     $$

7. **Instantaneous acceleration $\mathbf{a}(t)$:**

   * Limit of average acceleration as $\Delta t \to 0$, i.e., the derivative of velocity:

     $$
     \mathbf{a}(t) = \lim_{\Delta t \to 0} \frac{\mathbf{v}(t + \Delta t) - \mathbf{v}(t)}{\Delta t} = \frac{d\mathbf{v}}{dt} = \frac{d^2 \mathbf{r}}{dt^2}
     $$

```python
import numpy as np
import matplotlib.pyplot as plt

# Time array
t = np.linspace(0, 10, 1000)
dt = t[1] - t[0]

# Position function r(t), example: r(t) = 3t^2 + 2t + 1
def r(t):
    return 3*t**2 + 2*t + 1

# 1. Position vector for all t
pos = r(t)  # shape (N,)

# 2. Displacement vector Δr between consecutive time points
delta_r = pos[1:] - pos[:-1]  # shape (N-1,)

# 3. Average velocity v_avg = Δr / Δt
v_avg = delta_r / dt  # shape (N-1,)

# 4. Instantaneous velocity v(t) by numerical differentiation
v_inst = np.gradient(pos, dt)  # shape (N,)

# 5. Change in velocity Δv between consecutive points
delta_v = v_inst[1:] - v_inst[:-1]  # shape (N-1,)

# 6. Average acceleration a_avg = Δv / Δt
a_avg = delta_v / dt  # shape (N-1,)

# 7. Instantaneous acceleration a(t) by numerical differentiation of velocity
a_inst = np.gradient(v_inst, dt)  # shape (N,)

# Plotting results
plt.figure(figsize=(10, 10))

plt.subplot(4, 1, 1)
plt.plot(t, pos, label='Position r(t)')
plt.title('Position vs Time')
plt.ylabel('Position')
plt.legend()

plt.subplot(4, 1, 2)
plt.plot(t[:-1], v_avg, label='Average Velocity')
plt.plot(t, v_inst, '--', label='Instantaneous Velocity')
plt.title('Velocity vs Time')
plt.ylabel('Velocity')
plt.legend()

plt.subplot(4, 1, 3)
plt.plot(t[:-1], a_avg, label='Average Acceleration')
plt.title('Average Acceleration vs Time')
plt.ylabel('Acceleration')
plt.legend()

plt.subplot(4, 1, 4)
plt.plot(t, a_inst, '--', label='Instantaneous Acceleration')
plt.title('Instantaneous Acceleration vs Time')
plt.xlabel('Time')
plt.ylabel('Acceleration')
plt.legend()

plt.tight_layout()
plt.show()
```
