To understand **dynamics** using **Newton‚Äôs Laws**‚Äîparticularly **Newton‚Äôs Second Law** (*F = ma*)‚Äîwe can structure the **problem-solving process as a step-by-step algorithm**. This connects **forces** (dynamics) with the **motion** (kinematics) they produce.

## ‚úÖ **Sequence of Algorithm (Step-by-Step)** ‚Äì From Forces to Motion

### üî∑ Step 1: **Define the Object/System**

* **What?** Identify the object(s) whose motion you want to study.
* **Why?** Forces and motion are applied *only* to the system of interest.

### üî∑ Step 2: **Draw the Free Body Diagram (FBD)**

* **How?**

  * Represent the object as a dot or block.
  * Draw all external forces acting on the object:

    * Gravity (mg)
    * Normal force (N)
    * Friction (f)
    * Tension (T)
    * Applied force (F)
    * Spring force (Hooke's law: F = -kx)
* **Why?** Visualizes all forces affecting motion.

### üî∑ Step 3: **Choose Coordinate System**

* **What?** Decide which direction is +x, +y, etc.
* **Why?** Helps break vector forces into components (especially on slopes).

### üî∑ Step 4: **Apply Newton‚Äôs 2nd Law: $\vec{F}_{\text{net}} = m\vec{a}$**

* **How?**

  * Sum the vector components of all forces:

    $$
    \sum F_x = m a_x,\quad \sum F_y = m a_y
    $$
* **Why?** Connects force (cause) to acceleration (effect).

### üî∑ Step 5: **Solve for Acceleration**

* Use Newton‚Äôs second law to calculate acceleration $\vec{a}$.

### üî∑ Step 6: **Use Kinematic Equations to Find Position & Velocity**

* Once $\vec{a}$ is known, apply kinematic formulas:

  * $v = v_0 + at$
  * $x = x_0 + v_0t + \frac{1}{2}at^2$
  * $v^2 = v_0^2 + 2a(x - x_0)$
* For 2D/3D motion: apply the same equations in **each direction** (x, y, z).

### üî∑ Step 7: **Interpret Results**

* Analyze trajectory, time to reach a point, final velocity, etc.
* Check for limiting conditions like constant velocity (a = 0), equilibrium (F\_net = 0), etc.

## üîç Example

**Problem**: A 2 kg block is pushed with a 10 N force on a frictionless surface. What is its position after 3 seconds?

### Step-by-step:

1. **Object**: Block, m = 2 kg
2. **Forces**: Only applied force F = 10 N; surface is frictionless.
3. **FBD**: One horizontal force: F = 10 N
4. **Newton‚Äôs 2nd Law**:

   $$
   F = ma \Rightarrow 10 = 2a \Rightarrow a = 5\ \text{m/s}^2
   $$
5. **Kinematic Equation**: (initial velocity $v_0 = 0$)

   $$
   x = x_0 + v_0t + \frac{1}{2}at^2 = 0 + 0 + \frac{1}{2}(5)(3)^2 = 22.5\ \text{m}
   $$
