---

## 📘 Newton’s Third Law – Key Statement:

**"For every action, there is an equal and opposite reaction."**

If object A exerts a force $F$ on object B, then object B exerts a force $-F$ on object A.

---

# ⚙️ Algorithm 1: Collision Example – Train and Marble

### 🧠 Goal: Understand how Newton’s third law applies during collision.

---

### ✅ Input:

* Mass of marble $m_m$
* Mass of train $m_t$
* Force of interaction during collision $F$

---

### 🔁 Steps:

#### Step 1: Define the objects

```text
Object 1: Marble (mass = m_m)
Object 2: Train  (mass = m_t), where m_t >> m_m
```

#### Step 2: Apply Newton’s Third Law

```text
Force on marble by train = F
Force on train by marble = -F
```

#### Step 3: Use Newton’s Second Law for each object

```text
a_m = F / m_m      # acceleration of marble
a_t = -F / m_t     # acceleration of train (opposite direction)
```

#### Step 4: Take ratio of accelerations

```text
|a_m / a_t| = |(F / m_m) / (F / m_t)| = m_t / m_m
```

#### Step 5: Conclusion

```text
Since m_t >> m_m → acceleration of marble >> acceleration of train.
```

---

### 📌 Output:

* Force magnitudes are equal.
* Accelerations are vastly different due to mass difference.
* Train appears unaffected, but the same magnitude of force acts.

## 🧍 Algorithm 2: Standing Person – Ground Force & Gravity

### 🧠 Goal: Identify interaction pairs in contact and gravitational forces.

---

### ✅ Input:

* Mass of person $m_p$
* Acceleration due to gravity $g$
* Normal contact force $N$

---

### 🔁 Steps:

#### Step 1: Identify forces on person

```text
Gravity: F_gravity = m_p * g (downward)
Normal force: N (upward)
```

#### Step 2: Apply Newton's Third Law – Gravity Pair

```text
Earth exerts F_gravity on person (downward)
Person exerts F_gravity on Earth (upward)
```

#### Step 3: Apply Newton's Third Law – Normal Force Pair

```text
Ground exerts N on person (upward)
Person exerts N on ground (downward)
```

#### Step 4: Analyze Equilibrium

```text
If person is not accelerating vertically:
Net force = 0 → N = m_p * g
```

---

### 📌 Output:

* Interaction pairs:

  * Gravity: Earth ↔ Person
  * Normal: Ground ↔ Person
* All forces are mutual and on different objects.

## 🧮 Python Code to Simulate Newton’s 3rd Law (Collision Case)

```python
def simulate_collision(m_marble, m_train, force):
    a_marble = force / m_marble
    a_train = force / m_train
    print(f"Force on both: {force} N")
    print(f"Acceleration of Marble: {a_marble:.2f} m/s²")
    print(f"Acceleration of Train: {a_train:.6f} m/s²")
    print(f"Acceleration ratio (marble/train): {a_marble / a_train:.0f}:1")


simulate_collision(m_marble=0.05, m_train=50000, force=10)
```

## 📋 Summary

| Topic             | Key Idea                                              |
| ----------------- | ----------------------------------------------------- |
| Newton’s 3rd Law  | Forces come in equal and opposite pairs.              |
| Interaction Pairs | Always act on two **different** objects.              |
| Misconception     | Equal force ≠ equal effect (due to mass difference).  |
| Applications      | Collisions, gravity, walking, jumping, rocket motion. |
