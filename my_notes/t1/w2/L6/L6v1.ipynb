## 🔑 **Key Concepts to Convert into Algorithm**

* Object on a horizontal surface
* Applied force $F$
* Object moves at **constant speed**
* Kinetic friction $f_k$
* Normal force $N$
* Coefficient of kinetic friction $\mu_k$
* Contact area does **not** affect friction
* Friction force is **independent of velocity**

---

## 📘 **Physics Model Summary**

* $f_k = \mu_k \cdot N$
* For horizontal surface: $N = mg$ (if no vertical components)
* $F = f_k$ for constant velocity (Newton’s First Law: zero net force)

---

## 🔣 **Algorithm (Step-by-Step)**

---

### Step 0: **Input Parameters**

* Mass of object: `m` (kg)
* Coefficient of kinetic friction: `μ_k`
* Gravitational acceleration: `g = 9.8 m/s²`
* Optional: contact area `A` (to prove irrelevance)
* Optional: velocity `v` (to prove irrelevance)

---

### Step 1: **Compute Normal Force**

```plaintext
N = m * g
```

> Because it's on a horizontal surface, and no vertical force applied other than weight.

---

### Step 2: **Compute Kinetic Friction Force**

```plaintext
f_k = μ_k * N
    = μ_k * m * g
```

---

### Step 3: **Check Motion Condition**

```plaintext
If object moves with constant speed:
    Applied force F = f_k
Else:
    F > f_k ⇒ acceleration
    F < f_k ⇒ deceleration or object stops
```

---

### Step 4: **Display Output**

```plaintext
Output:
- Normal force: N
- Kinetic friction: f_k
- Required applied force F to maintain constant speed: f_k
```

---

## ✅ **Properties to Include**

* $f_k$ is **independent of contact area A**
* $f_k$ is **independent of velocity v**

You can simply state:

```plaintext
Friction force is independent of contact area A and velocity v.
```

---

## 🧪 Example

### Given:

* Mass `m = 10 kg`
* Coefficient of kinetic friction `μ_k = 0.3`

### Computation:

```plaintext
N = 10 * 9.8 = 98 N
f_k = 0.3 * 98 = 29.4 N
⇒ Required F = 29.4 N for constant velocity
```

---

## 🧠 Proof Outline

* Newton’s second law: $F - f_k = ma$
* For **constant velocity**: $a = 0 \Rightarrow F = f_k$
* Friction law: $f_k = \mu_k N$
* Surface area doesn’t affect $f_k$
* Experiments show $f_k$ stays constant with different velocities ⇒ supports velocity independence
