Here is the **step-by-step algorithm** to compute **gravitational acceleration (g)** at the **surface of the Earth** based on the text and the attached image.

---

### üîë **Key Equation**

The gravitational force between two masses:

$$
\vec{F}_{12} = -G \frac{m_1 m_2}{r_{12}^2} \hat{r}_{12}
$$

When on Earth‚Äôs surface:

$$
F = mg \Rightarrow g = \frac{GM_E}{R_E^2}
$$

---

### ‚úÖ **Algorithm: Step-by-Step Calculation of $g_{\text{Earth}}$**

#### **Step 0: Definitions (Initialize constants)**

* $G = 6.674 \times 10^{-11} \, \mathrm{N \cdot m^2/kg^2}$ ‚Äî Universal gravitational constant
* $M_E = 5.972 \times 10^{24} \, \mathrm{kg}$ ‚Äî Mass of the Earth
* $R_E = 6.371 \times 10^6 \, \mathrm{m}$ ‚Äî Radius of the Earth

---

#### **Step 1: Write Newton's Law of Universal Gravitation**

$$
F = G \frac{M_E m}{R_E^2}
$$

---

#### **Step 2: Equate gravitational force to weight**

$$
mg = G \frac{M_E m}{R_E^2}
$$

---

#### **Step 3: Cancel test mass $m$ from both sides**

$$
g = G \frac{M_E}{R_E^2}
$$

---

#### **Step 4: Plug in values**

$$
g = \frac{6.674 \times 10^{-11} \times 5.972 \times 10^{24}}{(6.371 \times 10^6)^2}
$$

---

#### **Step 5: Perform numerical calculation**

$$
g \approx \frac{3.986 \times 10^{14}}{4.058 \times 10^{13}} \approx 9.81 \, \mathrm{m/s^2}
$$

---

### üß™ Example in Code (Proof)

```python
# Gravitational acceleration at Earth's surface
G = 6.674 * 10**-11         # N¬∑m¬≤/kg¬≤
M_E = 5.972 * 10**24        # kg
R_E = 6.371 * 10**6         # m

g = G * M_E / R_E**2
print(f"Gravitational acceleration (g): {g:.2f} m/s¬≤")
```
